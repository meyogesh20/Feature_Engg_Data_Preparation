# Dealing with Categorical Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'D:\Data Science 2024\Study\ames_NO_Missing_Data.csv')
```

```python
df.isnull().sum().sort_values
```

* There is no missing values in the dataset

```python
df['MS SubClass'].head()
```

```python
# From dataSummary.txt 'MS SubClass' consists of categories represented by Numbers
# Lets convert those into the string so that we can apply get_dummies method
df['MS SubClass'] = df['MS SubClass'].apply(str)
```

```python
# Check the datatypes of all the features
df.select_dtypes(include='object') # string is of the object data type
```

```python
# Break the dataset into 2 parts
# 1. String Features
my_object_df = df.select_dtypes(include='object')
# 2. Numeric Features
my_numeric_df = df.select_dtypes(exclude='object')
```

```python
df_objects_dummies = pd.get_dummies(my_object_df,drop_first=True) #.astype(int)
df_objects_dummies
```

```python
final_df = pd.concat([my_numeric_df,df_objects_dummies],axis=1)
final_df
```

```python
# Find out the correlation of each feature with 'SalesPrice'
final_df.corr()['SalePrice'].sort_values()
```

### We can say that Overall Qual has a strong correlation with the Sale Price.
* Overall Qual attribute is based on the perspective of the human/person
* One just ask a question about the home and they have converted this rating into the Overall Quality
* But in case of prediction, who is going to gives the value of Overall Quality of the Home, who will be deciding it?
* It will become a pain point while developing a model.
